Para los números compuestos, cada número debe ser 'separado' en unidades, decenas y centenas. Una vez hecho esto se eligen de la lista numBasicos los nombres adecuados. Vamos a crear una función para este problema particular.

```python
def separaNumero(y):
    #Función que separa un número en unidades, decenas y centenas
    #Considero la división entera porque es más sencilla
    if y/100!=0: #La división entera, da enteros diferentes de cero para "múltiplos" de 100
        centenas=y/100
        y-=100*centenas #Reasigno y quito las centenas
        if y/10!=0:
            decenas=y/10
            y-=10*decenas
            unidades=y #Considerados todos los casos, sólo basta asignar las unidades
        else:
            decenas=0 
    else:
        centenas=0  
        if y/10!=0:
            decenas=y/10
            y-=10*decenas
            unidades=y
        else:
            decenas=0  
    unidades=y #Considerados todos los casos, sólo basta asignar las unidades
    return centenas, decenas, unidades   
```

Ahora podemos crear la funcion escribeNumero a partir de escribeNumeroBasico y separaNumero.

```python
def escribeNumero(z):
    #Escribe en palabras un numero entre 1 y 1000
    numero=""
    if z in range(1,21,1):  #Del 1 al 20 que ya sabemos escribir
        numero=str(escribeNumeroBasico(z))
    elif z in range(30,91,10):
        numero=str(escribeNumeroBasico(z)) #Del 30 al 90 decenas que ya sabemos escribir
    elif z==1000:
        numero="one hundred" #Añadiendo el caso extra
    else:
        (a,b,c)=separaNumero(z) #Primero separamos en numero
        if a!=0: #Consideramos la escritura de números entre 100 y 999: one+hundred+...
            numero=numero+str(escribeNumeroBasico(a))+" "+str(numBasicos[27]) #Agregando el hundred y cambiando a string
            if b==0: 
                numero=numero #Decenas vacías
                if c!=0: 
                    numero=numero+" "+str(escribeNumeroBasico(c)) #Unidades
                else:
                    numero=numero #Unidades vacías
            elif b!=0 and 10*b>=20: #Condicion para tener decenas arriba de 20 y hacer el thirty+...
                numero=numero+" "+str(escribeNumeroBasico(10*b)) #Decenas
                if c!=0:
                    numero=numero+" "+str(escribeNumeroBasico(c)) #Unidades
            else: #Si las decenas son menores 20, calculamos su número básico
                numero=numero+" "+str(escribeNumeroBasico(10*b+c))
        else: #Centenas vacías
            if 10*b>=20: #Condicion para tener decenas arriba de 20 y hacer el thirty+...
                numero=str(escribeNumeroBasico(10*b)) #Decenas
                if c!=0:
                    numero=numero+" "+str(escribeNumeroBasico(c)) #Unidades
            else: #Si las decenas son menores 20, calculamos su número básico
                numero=numero+" "+str(escribeNumeroBasico(10*b+c))
            
    return numero
```
